```python
import tarfile
```

```python
data_path_2019 = '/mimer/NOBACKUP/groups/globalpoverty1/albin_and_albin/raw_data_newest/2019/Gabon'
data_path_2020 = '/mimer/NOBACKUP/groups/globalpoverty1/albin_and_albin/raw_data_newest/2020/Gabon'
shard_path = '/mimer/NOBACKUP/groups/globalpoverty1/albin_and_albin/training_data_newest_lstm/Gabon_Shards'
shard_num = 0
max_idx = len(gabon2020)
for i in range(0, max_idx, 128):
    tar_shard = tarfile.open(f'{shard_path}/shard_{shard_num:06}.tar', 'w')
    
    if i + 128 >= max_idx:
        stop = max_idx
    else:
        stop = i + 128
    
    for i in range(i, stop):
        tar_shard.add(data_path_2019 + '/' + gabon2019[i])
        tar_shard.add(data_path_2020 + '/' + gabon2020[i])
    
    shard_num += 1
    print(f'shard {shard_num} done')
    tar_shard.close()
    
    
```
