# Metric Analysis: Landcover Prediction (boxplot)

```python
headers = ['recall', 'precision', 'iou', 'f1']
sub_headers = ['c1:Water', 'c2:Tree', 'c3:Flooded Vegetation', 'c4:Crops', 'c5:Bare Ground', 'c6:Rangeland', 'c7:Rural', 'c8:Urban']
cols = pd.MultiIndex.from_tuples([(header, subheader) for header in headers for subheader in sub_headers])

df_classification_metrics = pd.read_csv('/mimer/NOBACKUP/groups/globalpoverty1/albin_and_albin/metric_summary/landcover_prediction_esri_countries.csv', index_col=0)
df_classification_metrics.columns = cols
df_classification_metrics
```

### Recall

```python
#df_classification_metrics.boxplot(by ='day', column =['total_bill'], grid = False)
df_classification_metrics['recall'].boxplot(fontsize='7', rot=45)
plt.title('boxplots of country-wise recall scores, grouped by class')
plt.xticks(ha='right')
plt.show()
```
